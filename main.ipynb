# Consensus in Multi-agent Systems

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from functions import *
```

## Parameters

```python
N_UAVS = 10
RANDOM_SEED = 42
X_RANGE = (0, 100)
Y_RANGE = (20, 100)
N_ITERATIONS = 50
PROTOCOLS = ["arithmetic_mean", "geometric_mean", "harmonic_mean", "mean_of_order_2"]
TOPOLOGIES = ["all_to_all", "n_nearest_neighbors", "random"]
N_NEIGHBORS = 4
NEIGHBORS_STEP = 3
LINE = 100*'-'
DOUBLE_LINE = 100*'='
```

## UAVs initialization 

```python
uavs_list = uavs_initialization(N_UAVS, RANDOM_SEED, X_RANGE, Y_RANGE)
uav_plotting(uavs_list, X_RANGE, Y_RANGE, "Initial UAV Positions")
```

## Simulations by different protocols and topologies

### Results dataframe initialization

```python
pt_results = pd.DataFrame(columns=["protocol", "topology", "convergence iteration"])
```

### Simulations for arithmetic mean protocol and different topologies

```python
protocol = PROTOCOLS[0]

print(DOUBLE_LINE)
print(f"Simulations for {protocol} protocol")
print(DOUBLE_LINE)

for i, topology in enumerate(TOPOLOGIES):
    result, convergence_iteration = simulation(N_UAVS, uavs_list, N_ITERATIONS, protocol, topology, N_NEIGHBORS, RANDOM_SEED)

    print(f"{i+1}. Simulation using {protocol} protocol and {topology} topology")
    print(LINE)
    print(f"Convergence iteration: {convergence_iteration}")    

    uav_plotting(uavs_list, X_RANGE, Y_RANGE, "Initial and final UAV Positions", result[-1])
    plot_results(result, protocol, N_UAVS, N_ITERATIONS)

    new_row = pd.DataFrame({"protocol": [protocol], "topology": [topology], "convergence iteration": [convergence_iteration]})
    pt_results = pd.concat([pt_results, new_row], ignore_index=True)

    print(LINE)
```

### Results analysis by different protocols and topologies

```python
pt_results["convergence iteration"] = pd.to_numeric(pt_results["convergence iteration"], errors="coerce")

best_protocol_by_mean, worst_protocol_by_mean = results_dataframe_analysis(pt_results, "protocol", "convergence iteration", "mean", max_value=N_ITERATIONS)
best_protocol_by_median, worst_protocol_by_median = results_dataframe_analysis(pt_results, "protocol", "convergence iteration", "median", max_value=N_ITERATIONS)

best_topology_by_mean, worst_topology_by_mean = results_dataframe_analysis(pt_results, "topology", "convergence iteration", "mean", max_value=N_ITERATIONS)
best_topology_by_median, worst_topology_by_median = results_dataframe_analysis(pt_results, "topology", "convergence iteration", "median", max_value=N_ITERATIONS)

display_results("Results displayed by average and median convergence iteration for different protocols", "protocol",
                best_protocol_by_mean, worst_protocol_by_mean, best_protocol_by_median, worst_protocol_by_median,
                DOUBLE_LINE, LINE)

display_results("Results displayed by average and median convergence iteration for different topologies", "topology",
                best_topology_by_mean, worst_topology_by_mean, best_topology_by_median, worst_topology_by_median,
                DOUBLE_LINE, LINE)
```
